```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('data.csv',index_col=0)
data_y = pd.DataFrame([x/8 for x in data['Nt']])
data_X = pd.DataFrame([x/20 for x in data['(mm)']])
```

```python
model = LinearRegression()
model.fit(data_X[30:45],data_y[30:45]) #取第31到第45組數據來做線性回歸
predict = model.predict(data_X[30:45])
predict_2 = model.predict(data_X[18:51])
```

```python
plt.figure(figsize=(25,12))
plt.plot([data_X.values[data_y.values.argmax()],data_X.values[data_y.values.argmax()]],[0,data_y.values.max()],c='orange',linewidth=3.0,label=f'UTS={data_y.values.max():.6f}')
plt.plot([data_X.values[-1][0],data_X.values[-1][0]],[0,data_y.values[-1][0]],c='green',linewidth=3.0,label=f'Fracture strain={data_X.values[-1][0]:.6f}')
plt.plot(data_X.values[30:45],predict,c='black',linewidth=3.0)
plt.plot([x+0.002 for x in data_X.values[18:51]],predict_2,c='red',linewidth=3.0)
plt.scatter(data_X,data_y)
plt.xlabel('Strain',size=16)
plt.ylabel('Stress',size=16)
plt.legend()
plt.show()
```
